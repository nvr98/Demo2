```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.DataFrame(pd.read_csv('Social_Network_Ads.csv'))

x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values

len(x), len(y)
```

```python
#Training and test data

X_train , X_test ,Y_train , Y_test  = train_test_split(x,y , test_size=0.2 , random_state=42)

len(X_train) , len(X_test),len(Y_train),len(Y_test)
X_train
```

```python
#Since it is a classification problem the feature needs to be scaled to a uniform range of value

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Creating the logistic regression model

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 45)

classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
Y_test
```

```python
#Compairing the results of Classifier and origingal dataset 
np.concatenate([Y_pred.reshape(len(Y_pred),1),Y_test.reshape(len(Y_test),1)],1)
```

```python
classifier.predict(sc.transform([[30,100000]]))
```

```python
#Confusion matrix

from sklearn.metrics import confusion_matrix , accuracy_score

cm = confusion_matrix

print(cm(Y_test , Y_pred))

accuracy_score(Y_test , Y_pred)
```
